```python
import pandas as pd

data = {
    'Open': [100, 102, 101, 105],
    'Close': [102, 101, 106, 107]
}

df = pd.DataFrame(data)

def calculate_pnl(row):
    pnl = (row['Close'] - row['Open']) * 100
    return pnl

# Apply the function row-wise and create 'Daily PnL' column
df['Daily PnL'] = df.apply(calculate_pnl, axis=1)

print(df)



```
